```julia
dist = Chi(1)
```

```julia
plot(dist, lw = 3, label = "")
```

```julia
nu = [1, 2, 3, 4];
```

```julia
dists = [Chi(ii) for ii in nu]
```

```julia
typeof(dists)
```

```julia
plot(dists, 
       lw = 3, 
       label = "")
```

```julia
dists = Chisq(4)
plot(dists, lw = 3, lc = :blue, label = "χ-square")
dists2 = Chi(4)
plot!(dists2, lw = 3, lc = :red, label = "χ")
```
